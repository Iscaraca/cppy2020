```
d = {}

def menu():
    print("(1) Create list")
    print("(2) Insert item")
    print("(3) Update item")
    print("(4) Delete item")
    print("(5) Show all items")
    print("(6) Delete table")
    print("(0) Quit")

def createList():
    totalLists = int(input("How many lists to create: "))
    
    for _ in range(totalLists):
        listName = input("List name: ")
        d[listName] = []

    print(" ")

def insertItem():
    for key, value in d.items() :
        print(key)

    listAccess = input("Which list to access: ")
    listInsert = input("What value to insert: ")
    d[listAccess].append(listInsert)

    print(" ")

def updateItem():
    for key, value in d.items() :
        print(key)

    listAccess = input("Which list to access: ")
    listReplaceIndex = input("What index to replace: ")
    listReplaceValue = input("What value to replace with: ")
    d[listAccess][int(listReplaceIndex)] = listReplaceValue

    print(" ")

def deleteItem():
    for key, value in d.items() :
        print(key)

    listAccess = input("Which list to access: ")
    listReplaceIndex = input("What index to delete: ")
    del d[listAccess][int(listReplaceIndex)]

    print(" ")

def deleteTable():
    for key, value in d.items() :
        print(key)

    listAccess = input("Which list to delete: ")
    del d[listAccess]

    print(" ")

def showItems():
    for key, value in d.items() :
        print(key)

    listAccess = input("Which list to show: ")

    for i in d[listAccess]:
        print(i)
    
    print(" ")

choice = ''
while choice != '0':
    menu()
    choice = input()

    if choice == '1':
        createList()
    elif choice == '2': 
        insertItem()
    elif choice == '3':
        updateItem()
    elif choice == '4':
        deleteItem()
    elif choice == '5':  
        showItems()
    elif choice == '6':
        deleteTable()
print("Bye!")
```
