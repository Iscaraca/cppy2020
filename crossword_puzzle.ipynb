# Cryptic crosswords

```python
grid = [[-1, 1, -1, -1, 2, -1, 3],
       [4, 0, 0, 0, 0, 0, 0],
       [-1, 0, -1, -1, 0, -1, 0],
       [5, 0, 6, 0, 0, 7, 0],
       [0, -1, 0, -1, -1, 0, -1],
       [8, 0, 0, 0, 0, 0, 0],
       [0, -1, 0, -1, -1, 0, -1]]
```

```python
def display_crossword(grid):
    for row in grid:
        print(' '.join(map(str, row)).replace('-1', ' '))
```

```python
display_crossword(grid)
```

```python
list_of_clues = {
    'A4':  {'clue': 'Gets hot providing your bread and butter (7)', 'answer': 'TOASTER'},
    'A5':  {'clue': 'Card dealer (7)', 'answer': 'REFEREE'},
    'A8':  {'clue': 'He made others waltz (7)', 'answer': 'STRAUSS'},
    'D1':  {'clue': 'What have I got in my pocket? It lets all precious things pass (4)', 'answer': 'HOLE'},
    'D2':  {'clue': 'Dot that joins other dots to make a fish, ram, etc. (4)', 'answer': 'STAR'},
    'D3':  {'clue': 'I get a new ring every year (4)', 'answer': 'TREE'},
    'D5':  {'clue': 'Hour when you want this, but can’t do it (4)', 'answer': 'RUSH'},
    'D6':  {'clue': 'Not what you want to be under (4)', 'answer': 'FIRE'},
    'D7':  {'clue': 'A crossword, but not this one (4)', 'answer': 'EASY'}
}
```

```python
def display_clues(list_of_clues):
    for clue_no in list_of_clues:
        print(f"{clue_no}. {list_of_clues[clue_no]['clue']}")
```

```python
display_clues(list_of_clues)
```

```python
from copy import deepcopy

def solve(original_grid, list_of_clues):
    grid = deepcopy(original_grid)
    no_of_clues = len(list_of_clues)
    correct_answers = 0
    while correct_answers != no_of_clues:
        display_crossword(grid)
        print()
        display_clues(list_of_clues)
        print()
        
        clue_for_solving = input('Solve a clue: ')
        answer = input('Your answer (in caps): ')
        
        try:
            if answer == list_of_clues[clue_for_solving]['answer']:
                correct_answers += 1
                clue_no = int(clue_for_solving[1:])

                if clue_for_solving[:1] == 'A':                
                    for row in range(len(grid)):
                        for cell in range(len(grid)):
                            if original_grid[row][cell] == clue_no:
                                for index in range(len(answer)):
                                    grid[row][index + cell] = answer[index]

                elif clue_for_solving[:1] == 'D':
                    for row in range(len(grid)):
                        for cell in range(len(grid)):
                            if original_grid[row][cell] == clue_no:
                                for index in range(len(answer)):
                                    grid[row + index][cell] = answer[index]
                print('Well done!')

            else:
                print('Wrong, try again.')
        except:
            print('That clue number doesn\'t exist.')
            
    display_crossword(grid)
    print('You did it! Congrats!')
            
```

```python
solve(grid, list_of_clues)
```
